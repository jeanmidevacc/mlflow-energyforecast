```python
import pandas as pd 
import json
```

# Works on the energy consumption data 

```python
df_nationalconsumption = pd.read_csv("data/consommation-quotidienne-brute.csv", sep = ";")
df_nationalconsumption["timestamp"] = pd.to_datetime(df_nationalconsumption["Date - Heure"])
```

```python
df_nationalconsumption.head()
```

```python
# Focus on the data related to the electricity consumption
df_nationalconsumption_electricity = df_nationalconsumption[["timestamp","Statut - RTE","Consommation brute électricité (MW) - RTE"]]
df_nationalconsumption_electricity.columns = ["timestamp","datastatus","electricityconsumption"]
df_nationalconsumption_electricity["day"] = df_nationalconsumption_electricity["timestamp"].apply(lambda x: x.strftime("%Y-%m-%d"))
```

```python
# Aggregate the data to a daily scale
aggfunc = {
    "electricityconsumption" : {
        "dailyconsumption" : "sum",
        "count_points" : "count"
    }
}
df_nationalconsumption_electricity_daily = df_nationalconsumption_electricity.groupby(["day","datastatus"]).agg(aggfunc).reset_index()
df_nationalconsumption_electricity_daily.columns = ["day","datastatus","dailyconsumption","count_points"]

# Get knowledge on the day
df_nationalconsumption_electricity_daily["day"] = pd.to_datetime(df_nationalconsumption_electricity_daily["day"])
df_nationalconsumption_electricity_daily["weekday"] = df_nationalconsumption_electricity_daily["day"].apply(lambda x: x.weekday())
df_nationalconsumption_electricity_daily["week"] = df_nationalconsumption_electricity_daily["day"].apply(lambda x: x.isocalendar()[1])
df_nationalconsumption_electricity_daily["month"] = df_nationalconsumption_electricity_daily["day"].apply(lambda x: x.month)
df_nationalconsumption_electricity_daily["year"] = df_nationalconsumption_electricity_daily["day"].apply(lambda x: x.year)
df_nationalconsumption_electricity_daily.tail()
```

## Works on the weather data  (temperature)

```python
# collect information on the area/region (https://www.ined.fr/fr/tout-savoir-population/chiffres/france/structure-population/regions-departements/)
df_population = pd.read_csv("data/population.csv")

total_populations = df_population.iloc[-1]


columns_weightarea = []
for column in df_population.columns[1:]:
    overall = total_populations[column]
    new_column = "weigth_" + column
    
    df_population[new_column] = df_population[column] / overall
    
    columns_weightarea.append(new_column)
    
df_population = df_population[["area"] + columns_weightarea]
df_population
```

```python
# Prepare the weather data
df_regionalweather = pd.read_csv("data/temperature-quotidienne-regionale.csv", sep = ";")
df_regionalweather["day"] = pd.to_datetime(df_regionalweather["Date"])
df_regionalweather = df_regionalweather[["day","Région","TMin (°C)","TMax (°C)","TMoy (°C)"]]
df_regionalweather.columns = ["day","area","min_temperature","max_temperature","mean_temperature"]
df_regionalweather = pd.merge(df_regionalweather, df_population[["area"] + columns_weightarea], on = ["area"])

# Make quick average of the min, max and mean temperature of the region
df_nationalweather_nw = df_regionalweather.groupby(["day"]).mean()[["min_temperature","max_temperature","mean_temperature"]]
df_nationalweather_nw.columns = ["avg_min_temperature","avg_max_temperature","avg_mean_temperature"]

# Make a weighted average version (based on the population estimation in the region of 2008)
df_regionalweather_w = df_regionalweather[["day","area","min_temperature","max_temperature","mean_temperature","weigth_population_estimation2008"]]
for column in ["min_temperature","max_temperature","mean_temperature"]:
    df_regionalweather_w [column] = df_regionalweather_w [column] * df_regionalweather_w ["weigth_population_estimation2008"]

# Aggregate the weighted regional value
df_nationalweather_w = df_regionalweather_w.groupby(["day"]).sum()

# Drop some table
del df_nationalweather_w["weigth_population_estimation2008"]

# Rename some columns
df_nationalweather_w.columns = ["wavg_min_temperature","wavg_max_temperature","wavg_mean_temperature"]

df_nationalweather = pd.concat([df_nationalweather_nw,df_nationalweather_w], axis = 1)
df_nationalweather.reset_index(inplace = True)
```

```python
df_nationalweather.head()
```

## Final steps on the data preparation

```python
# Merge weather and electricity information
df_nationalconsumption_electricity_daily = pd.merge(df_nationalconsumption_electricity_daily,df_nationalweather, on =["day"])
```

```python
# Add the holiday
import holidays

#define the holiday object for france
holidays_france = holidays.France()

# Get the holiday for the day in France
df_nationalconsumption_electricity_daily["is_holiday"] = df_nationalconsumption_electricity_daily["day"].apply(lambda day: day in holidays_france)
# Convert boolean to 0,1
df_nationalconsumption_electricity_daily["is_holiday"] = df_nationalconsumption_electricity_daily["is_holiday"] * 1
# Set the index 
df_nationalconsumption_electricity_daily.set_index(["day"], inplace = True)
```

```python
# Select only the day that have 48 points of measurement (every 30 minutes)
df_nationalconsumption_electricity_daily = df_nationalconsumption_electricity_daily[df_nationalconsumption_electricity_daily["count_points"] == 48]
del df_nationalconsumption_electricity_daily["count_points"]
```

```python
# Save the data ready to use
df_nationalconsumption_electricity_daily.to_csv("data/rtu_data.csv")
```
